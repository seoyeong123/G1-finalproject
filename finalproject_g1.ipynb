### (1) Match words and meaning

```python
import random

# 단어 목록과 뜻 (예시)
words = {
    "vibrant": "활기찬",
    "ancient": "아주 오래된.",
    "scholarship": "장학금",
    "competition": "대회",
    "skeptics": "의심하는 사람",
    "donation": "기부"
    }


# 단어 목록과 뜻을 섞기
word_items = list(words.items())
random.shuffle(word_items)

# 단어와 뜻을 분리하여 섞기
english_words = [item[0] for item in word_items]
korean_words = [item[1] for item in word_items]
random.shuffle(korean_words)

# 학생들에게 단어와 뜻을 매칭시키는 함수
def word_matching_game(english_words, korean_words):
    score = 0
    for i, english_word in enumerate(english_words):
        print(f"Match the word: {english_word}")
        for j, korean_word in enumerate(korean_words, start=1):
            print(f"{j}. {korean_word}")

        try:
            answer = int(input("Your choice (enter the number): ").strip())
            if words[english_word] == korean_words[answer - 1]:
                print("Correct!\n")
                score += 1
            else:
                print(f"Incorrect. The correct answer is: {words[english_word]}\n")
        except (ValueError, IndexError):
            print(f"Invalid choice. The correct answer is: {words[english_word]}\n")

    print(f"You got {score} out of {len(english_words)} correct.")

# 게임 실행
word_matching_game(english_words, korean_words)
```

### (2) Practice spelling with hangman game

```python
import random

# 단어 목록
word_list = ["vibrant", "ancient", "scholarship", "competition", "skeptics", 'donation']
hints = {
    "vibrant": "Energetic",
    "ancient": "Very old.",
    "scholarship": "A sum of money or other aid granted to a student",
    "competition": "A contest for some prize",
    "skeptics": "A person who questions the authencity of something",
    "donation": "a gift as to a fund"
    }

def get_random_word(word_list):
    return random.choice(word_list)

def display_hangman(tries):
    stages = [  # Final state: head, torso, both arms, and both legs
                """
                   -----
                   |   |
                   O   |
                  /|\\  |
                  / \\  |
                       |
                --------
                """,
                # Head, torso, both arms, and one leg
                """
                   -----
                   |   |
                   O   |
                  /|\\  |
                  /    |
                       |
                --------
                """,
                # Head, torso, and both arms
                """
                   -----
                   |   |
                   O   |
                  /|\\  |
                       |
                       |
                --------
                """,
                # Head, torso, and one arm
                """
                   -----
                   |   |
                   O   |
                  /|   |
                       |
                       |
                --------
                """,
                # Head and torso
                """
                   -----
                   |   |
                   O   |
                   |   |
                       |
                       |
                --------
                """,
                # Head
                """
                   -----
                   |   |
                   O   |
                       |
                       |
                       |
                --------
                """,
                # Initial empty state
                """
                   -----
                   |   |
                       |
                       |
                       |
                       |
                --------
                """
    ]
    return stages[tries]

def play_hangman(word_list, hints):
    word = get_random_word(word_list)
    hint = hints[word]
    word_completion = "_" * len(word)
    guessed = False
    guessed_letters = []
    guessed_words = []
    tries = 6
    print("Let's play Hangman!")
    print(display_hangman(tries))
    print(f"Hint: {hint}")
    print(word_completion)
    print("\n")

    while not guessed and tries > 0:
        guess = input("Please guess a letter or word: ").lower()
        if len(guess) == 1 and guess.isalpha():
            if guess in guessed_letters:
                print("You already guessed the letter", guess)
            elif guess not in word:
                print(guess, "is not in the word.")
                tries -= 1
                guessed_letters.append(guess)
            else:
                print("Good job,", guess, "is in the word!")
                guessed_letters.append(guess)
                word_as_list = list(word_completion)
                indices = [i for i, letter in enumerate(word) if letter == guess]
                for index in indices:
                    word_as_list[index] = guess
                word_completion = "".join(word_as_list)
                if "_" not in word_completion:
                    guessed = True
        elif len(guess) == len(word) and guess.isalpha():
            if guess in guessed_words:
                print("You already guessed the word", guess)
            elif guess != word:
                print(guess, "is not the word.")
                tries -= 1
                guessed_words.append(guess)
            else:
                guessed = True
                word_completion = word
        else:
            print("Not a valid guess.")

        print(display_hangman(tries))
        print(word_completion)
        print("\n")

    if guessed:
        print("Congrats, you guessed the word! You win!")
    else:
        print(f"Sorry, you ran out of tries. The word was {word}. Maybe next time!")

# 게임 실행
play_hangman(word_list, hints)
```

```python
import gradio as gr

story = "https://raw.githubusercontent.com/MK316/Spring2024/main/DLTESOL/project/story01.txt"

# Define the comprehension questions
questions = [
    {"question": "1. What did the villagers believe about the Tree of Dreams?", "options": ["It was the oldest tree in the village.", "It could grant wishes if someone whispered their dreams to it.","It was a good place for children to play."], "answer": "It could grant wishes if someone whispered their dreams to it."},
    {"question": "2. What event prompted Mr. Lopez to encourage Maria to enter the national art competition?", "options": ["He saw her drawing a beautiful landscape.", "He saw her using leaves and coal pieces to create art.", "He received a letter from Maria's parents."], "answer": "He saw her using leaves and coal pieces to create art."},
    {"question": "3. What was Maria's wish when she visited the Tree of Dreams?", "options": ["To find a way to get professional art supplies for the competition.", "To show her best in the art competition.", "To be able to teach art to other children."], "answer": "To show her best in the art competition."},
    {"question": "4. Why did Maria use unconventional materials like leaves, petals, and coal pieces for her art?", "options": ["She thought they were better than regular art supplies.", "She was allergic to traditional art supplies.","Her family couldn’t afford to buy her art supplies.", "Her art teacher recommended these materials.", "She wanted to create a unique style."], "answer": "Her family couldn’t afford to buy her art supplies."},
    {"question": "5. What was the outcome of Maria's participation in the national art competition?","options":["She was disqualified for using unconventional materials.", "She received second place and a small scholarship.", "She won first place and a scholarship to any art school of her choice."], "answer": "She won first place and a scholarship to any art school of her choice."}
     ]
# Function to display the story and ask questions
def quiz(*answers):
    score = 0
    incorrect = []
    for i, question in enumerate(questions):
        if answers[i] == question["answer"]:
            score += 1
        else:
            incorrect.append(f"Question {i+1}: Correct answer is {question['answer']}")
    results = "\n".join(incorrect)
    return score, results

story_display = gr.Markdown(story)
inputs = [gr.Radio(label=q["question"], choices=q["options"]) for q in questions]
outputs = [gr.Textbox(label="Score"), gr.Textbox(label="Incorrect Answers")]

iface = gr.Interface(fn=quiz, inputs=inputs, outputs=outputs, description="Comprehension Quiz for 'The tree of Dreams'")
iface.launch(share=True)
```

### (1) Quiz: 5 sentences using participles in the story

```python
import gradio as gr

# The story "The Tree of Dreams" with blanks for cloze questions
story = "https://raw.githubusercontent.com/MK316/Spring2024/main/DLTESOL/project/story01.txt"

# Cloze questions
cloze_questions = [
    {"question": "Maria worked day and night, ______ (pour) her heart into her artwork.", "answer": "pouring"},
    {"question": "Most adults dismissed this as an old tale, ______ (mean) only to amuse children", "answer": "meant"},
    {"question": " ______ (overjoy) and grateful, Maria worked day and night.", "answer": "overjoyed"},
    {"question": "Maria continued to grow as an artist, always ______ (remember) the tree that once gave her hope", "answer": "remembering"},
    {"question": "The Tree of Dreams stood as a beacon of hope for everyone in Lila, ______ (remind) them that the oldest stories hold the deepest truths.", "answer": "reminding"}
    ]

# Function to handle the cloze quiz
def cloze_quiz(name, *answers):
    score = 0
    results = []
    for i, question in enumerate(cloze_questions):
        if answers[i].strip().lower() == question["answer"].lower():
            score += 1
            results.append(f"Question {i+1}: Correct\n")
        else:
            results.append(f"Question {i+1}: Incorrect, the correct answer is: {question['answer']}\n")
    result_text = f"* Name: {name}\n* Score: {score} out of {len(cloze_questions)}\n" + "\n".join(results)

    return result_text

# Add the input for the name
inputs = [gr.Textbox(label="Enter your name")] + [gr.Textbox(label=q["question"]) for q in cloze_questions]
outputs = [gr.Textbox(label="Results")]

iface = gr.Interface(fn=cloze_quiz, inputs=inputs, outputs=outputs, description="Cloze Quiz for 'The Tree of Dreams'")
iface.launch(share=True)
```

### (2) Check the understanding of participial construction

```python
import random

# 연습 문제 목록과 정답 (분사구문으로 변환된 문장)
exercises = [
    {
        "original": "Since she was tired, she went to bed early.",
        "participle": "Being tired, she went to bed early."
    },
    {
        "original": "After he finished his homework, he watched TV.",
        "participle": "Having finished his homework, he watched TV."
    },
    {
        "original": "When the boy saw the dog, he ran away.",
        "participle": "Seeing the dog, the boy ran away."
    },
    {
        "original": "Because the man was hungry, he ate a sandwich.",
        "participle": "Being hungry, the man ate a sandwich."
    },
]

# 게임 시작
def participle_explorer(exercises):
    random.shuffle(exercises)
    score = 0


    print("Your mission is to find the correct participle phrase for each sentence.")


    for i, exercise in enumerate(exercises, start=1):
        print(f"Challenge {i}:")
        print(f"Original sentence: {exercise['original']}")
        answer = input("What is the participle phrase? ").strip()
        if answer.lower() == exercise['participle'].lower():
            print("Correct! You found the hidden treasure!\n")
            score += 1
        else:
            print(f"Oops! That's not the right participle phrase. Keep exploring!\n")

    print(f"You found {score} treasures out of {len(exercises)}.")

# 게임 실행
participle_explorer(exercises)
```
